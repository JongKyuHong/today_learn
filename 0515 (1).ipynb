```
### 파일을 불러온다.
### nltk를 이용해서 생성(Text)
### 빈도를 확인
### 단어 빈도를 이용한 WordCloud를 만든다
### 이를 확인

doc_ko = open('news2.txt').read()
doc_ko[1:1000]
```

```
#plt.imshow(wc, interpolation='bilinear')
#plt.axis('off')
#plt.show()
from konlpy.tag import Okt ### Okt
t = Okt()
doc_nouns = t.nouns(doc_ko)
doc_nouns
ko = nltk.Text(doc_nouns, name="뉴스")
print(type(ko))
print(len(ko.tokens))

#Car_mask = np.array(Image.open("Draw_car1.png"))
```

```
most_fre = ko.vocab().most_common(50)
most_fre
```

```
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```
stop_words = ['훈련','보도','군','고','우리','청와대']

new_ko = [ ]
for one_word in ko:
  if one_word not in stop_words:
    new_ko.append(one_word)
print(new_ko)
```

```
new_ko = nltk.Text(new_ko, name="분노의 질주2")
plt.figure(figsize=(12,6))
new_ko.plot(50)
```

```
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```
Car_mask = np.array(Image.open("Draw_car1.png"))
```

```
data = new_ko.vocab().most_common(20)
#data2 = new_ko.vocab().most_common(1000)
```

```
wc = WordCloud(background_color='white',
 max_words=1000,
 mask=Car_mask,
 contour_width=3,
 contour_color="steelblue",
 font_path=path).generate_from_frequencies(dict(data))
```

```
plt.figure(figsize=(12,8))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
